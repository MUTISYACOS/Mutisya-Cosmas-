```python
#Data Cleaning
# Keep only the required columns
uae_pesticides = uae_pesticides[[
    'substance_id',
    'substance_name',
    'substance_status',
    'approval_date',
    'expiry_date',
    'risk_assessment',
    'substance_category',
    'as_is_group',
    'as_micro_org',
    'pesticide_residue_linked'
]]

# Display the first 5 rows
uae_pesticides.head()
```

```python
# Check Missing
uae_pesticides.isnull().sum()
```

```python
#Removing duplicates
uae_pesticides.drop_duplicates(inplace=True)
```

```python
#Data Shape after cleaning
print(uae_pesticides.shape)
```

```python
#Approval Period
uae_pesticides['approval_date'] = pd.to_datetime(
    uae_pesticides['approval_date'],
    dayfirst=True,
    errors='coerce'
)

uae_pesticides['expiry_date'] = pd.to_datetime(
    uae_pesticides['expiry_date'],
    dayfirst=True,
    errors='coerce'
)
# Calculate approval period in years
uae_pesticides['approval_period_years'] = (
    uae_pesticides['expiry_date'] - uae_pesticides['approval_date']
).dt.days / 365
average_period = uae_pesticides['approval_period_years'].mean()

print("Average approval period:", round(average_period, 2), "years")
```

```python
#Conditional Filtering
approved_efsa = uae_pesticides[
    (uae_pesticides['substance_status'] == 'Approved') &
    (uae_pesticides['risk_assessment'] == 'EFSA')
]

print("Number of approved substances assessed by EFSA:")
print(len(approved_efsa))
```

```python
# Display approved herbicides
approved_herbicides = uae_pesticides[
    (uae_pesticides['substance_status'] == 'Approved') &
    (uae_pesticides['substance_category'] == 'HB - Herbicide')
]

approved_herbicides[['substance_name', 'substance_category']]
```

```python
#Q4: # Count substances in each category
category_count = uae_pesticides.groupby('substance_category')['substance_id'].count()

category_count
```

```python
#Question 5: approval period in years
uae_pesticides['approval_period_years'] = (
    uae_pesticides['expiry_date'] -
    uae_pesticides['approval_date']
).dt.days / 365

# Sort approval periods
top10 = uae_pesticides.sort_values(
    by='approval_period_years',
    ascending=False
)

top10[['substance_name',
       'approval_period_years']].head(10)
```

```python
#Question 6 Filter and sort
efsa_approved = uae_pesticides[
    (uae_pesticides['substance_status'] == 'Approved') &
    (uae_pesticides['risk_assessment'] == 'EFSA')
]

efsa_approved = efsa_approved.sort_values(
    by='approval_period_years',
    ascending=False
)

efsa_approved[['substance_name',
               'approval_period_years',
               'risk_assessment']].head(10)
```

```python
#Q7. Mixed Condition + Grouping
# Average approval period by risk assessment
risk_average = uae_pesticides.groupby(
    'risk_assessment'
)['approval_period_years'].mean()

risk_average = risk_average.sort_values(
    ascending=False
)

risk_average
```
